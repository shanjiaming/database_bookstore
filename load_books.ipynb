```python
from fe.access.book_sqlite import *
# use small dataset
bookdb = BookDB(False)
booklist = bookdb.get_book_info(0, bookdb.get_book_count())

from pymongo import MongoClient

# create a MongoClient object and connect to the database
client = MongoClient()
db = client['bookstore']

# create a collection named 'books'
collection = db['books']

# create a list of Book objects
books = booklist

# insert the list of books into the collection
collection.insert_many([book.__dict__ for book in books])
```

```python
cursor = collection.find()

# print out the documents
print(cursor[1])
```
